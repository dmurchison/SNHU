```python
import matplotlib.pyplot as plt

# Displacement (m) data
displacement = [0.25, 0.50, 0.75, 1.00, 1.25, 1.50, 1.75, 2.00]

# Corresponding time (s) data
time = [0.7775, 1.555, 2.332, 3.110, 3.887, 4.665, 5.442, 6.220]

# Create the plot
plt.plot(displacement, time, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Displacement (m)')
plt.ylabel('Time (s)')
plt.title('Time vs Displacement for a Toy Car')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Time (s) data
time = [1, 2, 3, 4, 5, 6, 7, 8]

# Corresponding velocity (m/s) data
velocity = [0.3215, 0.3215, 0.3215, 0.3215, 0.3215, 0.3215, 0.0, 0.0]

# Create the plot
plt.plot(time, velocity, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Velocity vs Time for a Toy Car')

# Draw a vertical line from the x-axis at the point time = 2 seconds so that it intersects
# the line representing the velocity of the car.
plt.axvline(x=2, color='r', linestyle='--')
# Draw a second vertical line from the x-axis at the point time = 4 seconds so that it
# intersects the line representing the velocity of the car.
plt.axvline(x=4, color='r', linestyle='--')
# Calculate the area represented by the rectangle enclosed by the two vertical lines
# you just drew, the line for the velocity of the car, and the x-axis.
plt.fill_between([2, 4], 0, 0.3215, color='r', alpha=0.5)

# To calculate the area of this rectangle, you need to multiply the the value of the time interval between the two
# vertical lines by the value of the velocity of the car. This is known as calculating the area under the curve.
# The area under the curve represents the displacement of the car during the time interval between the two vertical lines.
area = 2 * 0.3215
print('The area under the curve is:', area) # saying the distance traveled by the car in 2 seconds, is 0.643 meters

# Show the plot
plt.grid(True)
plt.show()
```
