```python
import matplotlib.pyplot as plt

# Force (N) data
force = [0.1862, 0.2476, 0.3078, 0.3599, 0.4276]

# Corresponding acceleration (m/s²) data
acceleration = [0.4125, 0.7333, 0.9296, 1.0, 1.0909]

# Create the plot
plt.plot(force, acceleration, marker='o', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Force (N)')
plt.ylabel('Acceleration (m/s²)')
plt.title('Force vs Acceleration')

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Force (N) data
force = np.array([0.1862, 0.2476, 0.3078, 0.3599, 0.4276])

# Corresponding acceleration (m/s²) data
acceleration = np.array([0.4125, 0.7333, 0.9296, 1.0, 1.0909])

# Perform linear regression to find the slope and intercept
slope, intercept = np.polyfit(force, acceleration, 1)

# Print the slope (which represents the mass of the system)
print("Slope of the line (mass of the system):", slope, "kg")

# Create the plot
plt.plot(force, acceleration, marker='o', linestyle='-', color='b', label='Data points')
plt.plot(force, slope * force + intercept, color='r', label='Best-fit line')
plt.xlabel('Force (N)')
plt.ylabel('Acceleration (m/s²)')
plt.title('Force vs Acceleration')
plt.legend()
plt.grid(True)
plt.show()
```
